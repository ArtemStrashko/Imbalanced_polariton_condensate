```python
import numpy as np
import time
import pickle
import matplotlib.pyplot as plt
from pylab import figure, plot, show, savefig
import sys
import os.path
from finite_T_min_init import *
from model_parameters import *
plt.rcParams.update({'font.size': 15})
```

## Set up physical parameters and run precalcs

```python
phys_params = give_parameters()
phys_params['target particle real space density n_0'] = 0.4
phys_params['photon cut-off freq'] = 1.55

extra_param = setting_up_grids__and_k_vect_len(phys_params)
phys_params.update(extra_param)

cs = cosine(phys_params["k-grid"])

attract_potential = Coulomb_modified(phys_params) 
phys_params["cosine"] = cs
phys_params["Coulomb interaction in real space"] = attract_potential

k = phys_params["k-grid"]
zero_ind = phys_params['zero momentum index']
num_of_vars = phys_params['number of variables']
```

## Define normal and full energy functions

```python
En_normal = lambda var_s: vMFFE_normal(var_s, phys_params, q)
E_coh = lambda var_s: vMFFE(var_s, phys_params, q)
```

## Define Q-vector array covering both small and large values

```python
q_array_1 = np.linspace(-2e-4, 1e-4, 41)[:-1]
q_array_2 = np.logspace(-4, -1, 20)[:-1]
q_array_3 = np.linspace(0.1, 2.0, 21)

q_array = np.append(q_array_1, q_array_2)
q_array = np.append(q_array, q_array_3)
q_array = np.append(q_array, q_array[::-1][1:])
```

## Set up zero arrays for collecting results

```python
Eq = np.zeros([len(q_array)])
solution_q = np.zeros([len(q_array), num_of_vars])
```

## Run minimization

```python
for i in range(len(q_array)):
    
    start_time = time.time()  
    
    q = q_array[i]
    other = { 'Is coherent?' : True, 'number of minimization repetitons' : 7, 'q-vector' : q }

    # coherent solution with memory 
    if i == 0:
        x_norm = initial_conditions_normal(phys_params, q)
        init_with_mem = initial_conditions_coherent(phys_params, q, x_norm)
    else:
        init_with_mem = result_with_mem[1]
    result_with_mem = get_new_minimum(init_with_mem, phys_params, other)
    Eq[i] = result_with_mem[0]
    solution_q[i,:] = result_with_mem[1]

    
        
    print('step ' + str(i+1) + ' out of ' + str(len(q_array)))
    print('E = ' + str(Eq[i]))
    print('q is ' + str(q_array[i]))
    print("--- %s seconds ---" % (time.time() - start_time) + '\n \n')  
```
